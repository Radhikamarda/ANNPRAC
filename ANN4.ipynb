```python
import numpy as np
from sklearn.datasets import make_classification

# Generate a random dataset with two classes
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                            n_informative=2, random_state=42, n_clusters_per_class=1)

import matplotlib.pyplot as plt

# Initialize the weights and bias
w = np.random.rand(2)
b = np.random.rand()

# Set the learning rate
lr = 0.1

# Define the activation function
def activate(x):
    return np.where(np.dot(w, x) + b > 0, 1, -1)

# Train the perceptron

for i in range(100):
    # Randomly select a data point
    index = np.random.randint(len(X))
    x = X[index]
    y_true = y[index]
    
    # Compute the predicted label
    y_pred = activate(x)
    
    # Update the weights and bias if the prediction is incorrect
    if y_pred != y_true:
        w = w + lr * y_true * x
        b = b + lr * y_true

# Plot the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y)
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = np.array([activate([x1, x2]) for x1, x2 in zip(xx.ravel(), yy.ravel())])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k')
plt.show()
    

```
