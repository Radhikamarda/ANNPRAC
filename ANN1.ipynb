```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha*x, x)

x= np.linspace(-10, 10, num= 100)
```

```python
plt.plot(x, sigmoid(x), label="sigmoid")
plt.xlabel("X")
plt.ylabel("Output")
plt.title("Activation Function")
plt.show()
```

```python
plt.plot(x, tanh(x), label="tanh")
plt.xlabel("X")
plt.ylabel("Output")
plt.title("Activation Function")
plt.grid()
plt.show()
```

```python
plt.plot(x, relu(x), label="relu")
plt.xlabel("X")
plt.ylabel("Output")
plt.title("Activation Function")
plt.show()
```

```python
plt.plot(x, leaky_relu(x), label="leaky_relu")
plt.xlabel("X")
plt.ylabel("Output")
plt.title("Activation Function")
plt.show()
```

```python
plt.plot(x, linear(x), label="linear")
plt.xlabel("X")
plt.ylabel("Output")
plt.title("Activation Function")
plt.grid()
plt.show()
```
